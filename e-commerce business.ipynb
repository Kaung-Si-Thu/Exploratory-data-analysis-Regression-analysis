```python
import duckdb
import csv
import os

# Path to DuckDB file
db_file_path = "/content/eda_sample_data.db"

con = duckdb.connect(database=db_file_path, read_only=True)

# the table to export
table_name = "eda_sample_data.store_data"

# directory to store CSV files
output_dir = "/content/drive/My Drive/exported_csv"
os.makedirs(output_dir, exist_ok=True)

# Export data from the specified table to CSV
output_file = os.path.join(output_dir, f"{table_name.replace('.', '_')}.csv")
with open(output_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)

    # Fetch column names
    columns = [col[0] for col in con.execute(f"SELECT * FROM {table_name} LIMIT 0").description]
    writer.writerow(columns)

    # Fetch all rows from the table
    cursor = con.execute(f"SELECT * FROM {table_name}")
    rows = cursor.fetchall()
    # Write rows to CSV file
    writer.writerows(rows)

# Close the connection
con.close()

print(f"CSV file for table {table_name} exported successfully.")
```

```python
import duckdb
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Path to DuckDB file
db_file_path = "/content/eda_sample_data.db"

# Connect to DuckDB
con = duckdb.connect(database=db_file_path, read_only=True)

# Specify the table to explore
table_name = "eda_sample_data.order_data"

# Fetch data from the table into a pandas DataFrame
df = con.execute(f"SELECT * FROM {table_name}").fetchdf()

# Display basic information about the DataFrame
print("DataFrame Info:")
print(df.info())

# Display summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Visualize distributions
plt.figure(figsize=(10, 6))
sns.histplot(df["total_price"], kde=True)
plt.title(f"Distribution of total_price in {table_name}")
plt.xlabel("Total Price")
plt.ylabel("Frequency")
plt.show()

# Explore relationships between variables

plt.figure(figsize=(10, 6))
sns.scatterplot(x="order_time", y="total_price", data=df)
plt.title(f"Relationship between order_time and total_price in {table_name}")
plt.xlabel("Order Time")
plt.ylabel("Total Price")
plt.xticks(rotation=45)
plt.show()

# Close the connection
con.close()
```
